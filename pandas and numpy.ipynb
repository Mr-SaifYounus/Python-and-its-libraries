```python
import pandas as pd

def add_new_column(df, column_name, values):
    if column_name not in df.columns:
        df[column_name] = values
    else:
        raise ValueError(f"Column '{column_name}' already exists in DataFrame")
    return df

data = {"Name": ["Amir", "Saif", "Younus"],
        "Age": ["25", "30", "35"]}
df = pd.DataFrame(data=data)

print("Original DataFrame:")
print(df)
print("="*50)

new_column_name = "Country"
new_column_values = ["Pakistan", "India", "Bangladesh"]

df = add_new_column(df, new_column_name, new_column_values)

print("Updated DataFrame:")
print(df)
```

```python
#input="aabcccccaaa"
#output="a2b1c5a3"
def compress_string(s): #defining the function
    compressed = []     #defining the dict
    count = 1           #starting the count
    for i in range(1, len(s)):  #loop
        if s[i] == s[i-1]:      #if the next element is same as previous one
            count += 1          #count increment by 1
        else:
            compressed.append(s[i-1] + str(count)) #append the pervious element by count
            count = 1                              #resetting count to 1
    compressed.append(s[-1] + str(count))          #for last element and string count 
    return ''.join(compressed)                     #join the elements by .join
Input_string= input("Enter a string: ")
print("iNPUT sTRING :", Input_string)
Output_str = compress_string(Input_string)
print("CompresseD String:", Output_str)
```

```python
#Implement a function that performs a vectorized operation to calculate the Manhattan distance between points in an N-dimensional space, using NumPy arrays.

def manhattan_distance(a,b):
    return np.sum(np.abs(a-b),axis=-1)
c=np.array([[1,2,3],[4,5,6]])
d=np.array([[7,8,9],[10,11,12]])
distance1=manhattan_distance(c,d)  
print(distance1)
```

```python
#Generate a signal composed of multiple sine waves with different frequencies. Use NumPys Fast Fourier Transform (FFT) to analyze the frequency components of the signal.
t=np.linspace(0,1,500)
print(t)
x=np.sin(2*np.pi*5*t)+np.sin(2*np.pi*10*t)
fft1=np.fft.fft(x)
print("="*50)
print(fft1)
print("="*50)
freq=np.fft.fftfreq(len(x),d=t[1]-t[0])
print(freq)
```
